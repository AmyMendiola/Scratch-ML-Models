```python
import numpy as np
import sklearn
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 

#load dataset
from sklearn.datasets import load_boston
boston_dataset = load_boston()
print(boston_dataset.keys())
print(boston_dataset.DESCR)

#make dataframe
boston = pd.DataFrame(boston_dataset.data, columns=boston_dataset.feature_names)
boston['MEDV'] = boston_dataset.target
print(boston)

#check for null values
print(boston.isnull().sum())

#split data
from sklearn.model_selection import train_test_split
X = boston.to_numpy()
X = np.delete(X, 13, 1)
y = boston['MEDV'].to_numpy()
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=5)
X_train=np.array(X_train, dtype=float)
X_test=np.array(X_test, dtype=float)
y_train=np.array(y_train, dtype=float)
y_test=np.array(y_test, dtype=float)
# print(X_train.shape)
# print(X_test.shape)
# print(y_train.shape)
# print(y_test.shape)
```

```python
learn_rate=0.000001
iterations=1000000

#fit the data
samples, features = X_train.shape
weights = np.zeros(features)
bias = 0

for _ in range(iterations):
    y_pred = np.dot(X_train, weights) + bias
    #calc gradients
    error = y_pred-y_train
    dw = (1/samples) * np.dot(X_train.T, error)
    db = (1/samples) * np.sum(error)
    #update weights
    weights = weights - (learn_rate * dw)
    bias = bias - (learn_rate * db)

#calculate the predicted y from X_test set
y_pred = np.dot(X_test, weights) + bias

#print results
print('\nScratch weight')
print(weights) #weights

def loss_funct(y_test, y_pred):
    return np.mean((y_test-y_pred)**2)

#calc accuracy (loss function)
t_samples, t_features = X_test.shape
print('loss function: ' + str(loss_funct(y_test,y_pred)))

##Compare to sklearn##
from sklearn.linear_model import LinearRegression
lin_model = LinearRegression()
lin_model.fit(X_train, y_train)
print('\n*COMPARISON*\n')
print('sklearn weights')
print(lin_model.coef_) #weights
y_pred_compare = lin_model.predict(X_test)
print('Comparison loss function: ' + str(loss_funct(y_test,y_pred_compare)))
```
